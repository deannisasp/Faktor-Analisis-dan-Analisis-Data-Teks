# Reliabilitas

```python
def cronbach_alpha(df):
    # 1. Transform the df into a correlation matrix
    df_corr = df.corr()

    # 2.1 Calculate N
    # The number of variables equals the number of columns in the df
    N = df.shape[1]

    # 2.2 Calculate R
    # For this, we'll loop through the columns and append every
    # relevant correlation to an array calles "r_s". Then, we'll
    # calculate the mean of "r_s"
    rs = np.array([])
    for i, col in enumerate(df_corr.columns):
        sum_ = df_corr[col][i+1:].values
        rs = np.append(sum_, rs)
    mean_r = np.mean(rs)

   # 3. Use the formula to calculate Cronbach's Alpha
    cronbach_alpha = (N * mean_r) / (1 + (N - 1) * mean_r)
    return cronbach_alpha
```

```python
data_reliabilitas_A = datasum.loc[0:54,'A1':'A10']
data_reliabilitas_A
```

```python
cronbach_alpha(data_reliabilitas_A)
```

Karena cukup tinggi nilai 0,788 hasili dari nilai cronbach, artinya Reliabilitas kuesioner kita cukup reliable (cukup baik) dan dapat digunakan. Sehingga kuesioner dapat digunakan untuk pengumpulan data untuk variable A

```python
data_reliabilitas_B = datasum.loc[0:54,'B1':'B10']
data_reliabilitas_B
```

# Hipotesis Testing

```python
from scipy.stats import ttest_1samp

# Hipotesis
print("Hipotesis :")
print("H0 : Beta 1 = 0 --> variabel tidak berpengaruh secara signifikan")
print("H1 : Beta 1 tidak sama dengan 0 --> variabel berpengaruh secara signifikan")
print("\n")

# Menentukan tingkat signifikansi (alpha)
alpha = 0.05
print("Menentukan tingkat signifikansi (alpha) :")
print("alpha =", alpha)
print("\n")

#Menentukan kriteria pengujian
print("Menentukan kriteria pengujian :")
print("Jika |Thit| > Ttabel atau p-value < 0.05 maka H0 ditolak")
print("\n")

# Data yang sesuai dari kolom 'A5' dalam DataFrame data_FA
data_A5 = data_FA['A5']

# Nilai hipotesis nol (H0)
null_hypothesis_mean = 0  # Ganti dengan nilai yang sesuai sesuai dengan hipotesis nol Anda

# Melakukan uji t-test satu sampel
t_statistic, p_value = ttest_1samp(data_A5, null_hypothesis_mean)
print("Menentukan uji statistik T :")
print("T-statistic:", t_statistic)
print("P-value:", p_value)
print("\n")

# Menentukan apakah hipotesis nol ditolak atau tidak
print("Kesimpulan :")
if p_value < alpha:
    print("H0 ditolak, artinya variabel berpengaruh secara signifikan")
else:
    print("Gagal tolak H0 artinya variabel tidak berpengaruh secara signifikan")
```

Mengapa menggunakan uji T? Karena standar deviasi populasi tidak diketahui, oleh karena itu menggunakan uji T. Variabel yang diuji adalah variabel A5 (hasil dari analisis FA) untuk mengetahui apakah variabel ini berpengaruh secara signifikan. Dan berdasarkan hasil pengujian di atas, dapat diketahui bahwa variabel A5 berpengaruh secara signifikan

```python
#Mengubah ke dalam bentuk lowercase, menghapus tanda baca, mengubah angka menjadi kalimat, dan menghapus spasi pada awal dan akhir kalimat
def preprocess_text(kalimat):
    lower_case = kalimat.lower()
    hasil = re.sub(r"\d+", "",lower_case)
    hasil = hasil.translate(str.maketrans('','',string.punctuation))
    hasil = hasil.strip()

    return hasil

df['Cleaned'] = df['C'].apply(preprocess_text)
df
```

```python
#Tokenized
def tokenize_text(kalimat):
    tokens = nltk.tokenize.word_tokenize(kalimat)
    return tokens

df['Tokenized'] = df['Cleaned'].apply(tokenize_text)
df
```

```python
def combine_words(word_list):
    all_words = []
    for word in word_list:
        all_words += word
    return all_words

def count_topwords(all_words):
    counts = dict()
    for word in all_words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    word_count = pd.DataFrame([counts])
    word_count_transposed = word_count.T.reset_index()
    word_count_transposed.columns = ['Kata','Frekuensi']
    word_count_sorted = word_count_transposed.sort_values("Frekuensi",ascending = False)
    word_count_sorted
    return word_count_sorted[:20]
```

```python
reviews = df['Tokenized']
reviews_words = combine_words(reviews)
reviews_words[:10]
```
